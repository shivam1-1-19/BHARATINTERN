```python
#Lowercasing:
#Convert all text to lowercase to ensure uniformity.
new_df['reviewText']=new_df['reviewText'].str.lower()
new_df['reviewText']
```

```python
#Removing Punctuation and Special Characters:
#Remove punctuation marks and any non-alphanumeric characters that do not add significant meaning.
import string

def remove_special_characters(reviewText):
    return ''.join(char for char in reviewText if char.isalnum() or char.isspace())

new_df['reviewText'] = new_df['reviewText'].apply(remove_special_characters)
new_df['reviewText']
```

```python
#Tokenization:
#Split the text into individual words or tokens.
def tokenize_text(reviewText):
    return reviewText.split()

new_df['reviewText']= new_df['reviewText'].apply(tokenize_text)
new_df['reviewText']
```

```python
#Stopword Removal:
#Remove common words like "the," "is," "a," etc., 
#known as stopwords, as they do not add significant meaning to sentiment analysis.
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def remove_stopwords(reviewText):
    return [word for word in reviewText if word not in ENGLISH_STOP_WORDS]

new_df['reviewText'] = new_df['reviewText'].apply(remove_stopwords)
new_df['reviewText']
```

```python
#Handling Negations:
#Consider handling negations (e.g., "not good")
#by marking the negation words or converting them to 
#their antonyms. We'll use a basic function to handle negations.\

def handle_negations(reviewText):
    negation_flag = False
    result = []
    
    for word in reviewText:
        if word in ['not', 'no', 'n\'t', 'never']:
            negation_flag = True
        else:
            if negation_flag:
                word = 'not_' + word  # Mark the word as a negation
                negation_flag = False
            result.append(word)
    
    return result

new_df['reviewText'] = new_df['reviewText'].apply(handle_negations)
new_df['reviewText']
```

```python
#Create a Sentiment Dictionary:
#Build a dictionary that contains emoticons or
#special words as keys and their corresponding sentiment polarities as values. 
sentiment_dict = {
    ":)": "positive",
    ":(": "negative",
    ":-)": "positive",
    ":-(": "negative",
}
#Assign Sentiment Labels:
def assign_sentiment_labels(reviewText, sentiment_dict):
    labeled_tokens = []

    for token in reviewText:
        if token in sentiment_dict:
            sentiment_label = sentiment_dict[token]
            labeled_tokens.append(f"{sentiment_label}_{token}")
        else:
            labeled_tokens.append(token)

    return labeled_tokens

new_df['reviewText'] = new_df['reviewText'].apply(lambda x: assign_sentiment_labels(x, sentiment_dict))
new_df['reviewText']
```

```python
#Numerical Values:Numerical Values:
import re
# Ensure the 'text' column is of string data type
new_df['reviewText'] = new_df['reviewText'].astype(str)

def remove_numerical_values(reviewText):
    return re.sub(r'\d+', '', reviewText)

new_df['reviewText'] = new_df['reviewText'].apply(remove_numerical_values)
new_df['reviewText']
```

```python
#URL  remove:
def remove_urls(reviewText):
    return re.sub(r'http\S+|www\S+|https\S+', '', reviewText)

new_df['reviewText'] = new_df['reviewText'].apply(remove_urls)
new_df['reviewText']
```

****

```python
#Removing Short Words:

import re

def remove_short_words(reviewText, min_word_length=3):
    # Use regex to remove words with lengths less than min_word_length
    return re.sub(r'\b\w{1,' + str(min_word_length - 1) + r'}\b', '', reviewText)

# Example usage:

new_df['reviewText'] =new_df['reviewText'].apply(remove_short_words)
new_df['reviewText']
```

```python
from sklearn.utils import resample
# Assuming you have a DataFrame 'new_df' with a column 'sentiment'

# Separate positive and negative classes
positive_samples = new_df[new_df['sentiment'] == 'POSITIVE']
negative_samples = new_df[new_df['sentiment'] == 'NEGATIVE']
# Upsample the negative class to match the size of the positive class
negative_samples_upsampled = resample(negative_samples, replace=True, n_samples=len(positive_samples), random_state=42)
# Combine the upsampled negative class with the positive class
balanced_df = pd.concat([positive_samples, negative_samples_upsampled])

x = balanced_df['reviewText']
y = balanced_df['sentiment']
x.shape
y.shape
#x---> previous reviewtext + new negative reviewText
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
# Step 2: Milestone 1 - Building baseline model with Logistic Regression and cross-validation
vectorizer = CountVectorizer()  # Adjust the max_features parameter as needed
X_transformed = vectorizer.fit_transform(x)
#X_transformed -- (Vectorized X)
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.4, random_state=42)

```

```python
baseline_model = LogisticRegression(max_iter=1000)

# Hyperparameter tuning
param_grid = {'C': [0.1, 1]}
grid_search = GridSearchCV(baseline_model, param_grid, cv=3)
grid_search.fit(X_train, y_train)

baseline_model = grid_search.best_estimator_

# Model evaluation
baseline_predictions = baseline_model.predict(X_test)  # Evaluate on a smaller sample
print("Baseline Model:")
print(classification_report(y_test, baseline_predictions))
```
